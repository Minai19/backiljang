```python
# 데이터를 '지역' 열로 그룹화하고 각 그룹의 학교 수와 체육관 개방 학교 수를 계산
grouped = info_hs.groupby('지역')
total_schools_per_region = grouped.size()
gyms_open_per_region = grouped['체육관 개방여부'].apply(lambda x: (x == '유').sum())

# 지역별 체육관 개방 비율 계산
gym_open_ratio_per_region = gyms_open_per_region / total_schools_per_region

# 스택형(bar) 그래프 그리기
plt.figure(figsize=(12, 6))
p1 = plt.bar(total_schools_per_region.index, total_schools_per_region, color='skyblue', label='전체 학교 수')
p2 = plt.bar(gyms_open_per_region.index, gyms_open_per_region, color='lightgreen', label='체육관 개방 학교 수')
plt.ylabel('학교 수')
plt.xlabel('지역')
plt.title('서울시 자치구별 고등학교 수 및 체육관 개방 학교 비율')
# x축 라벨을 공백 뒤의 글자만 표시
plt.xticks(range(len(total_schools_per_region)), [label.split()[-1] for label in total_schools_per_region.index], rotation=45)
plt.legend((p1[0], p2[0]), ('전체 학교 수', '체육관 개방 학교 수'))

# 막대 그래프 위에 숫자로 체육관 개방 비율 나타내기
for i, v in enumerate(gym_open_ratio_per_region):
    plt.text(i, v * total_schools_per_region[i], f'{v:.2f}', ha='center', va='bottom')


# 그래프 출력
plt.show()
```
